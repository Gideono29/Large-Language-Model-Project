```python
# Import libraries
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import importlib.metadata
from packaging import version

# Check to see if the correct version of bitsandbytes is installed
try:
  bitsandbytes_version = importlib.metadata.version("bitsandbytes")
  print(f"bitsandbytes version: {bitsandbytes_version}")
  if version.parse(bitsandbytes_version) < version.parse("0.39.0"):
      print("bitsandbytes version is too old. Please update to 0.39.0 or higher.")
except importlib.metadata.PackageNotFoundError:
  print("bitsandbytes is not installed. Please install it.")

# Model details
model_name = "meta-llama/Meta-Llama-3-8B"

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load the model with 4-bit quantization onto the GPU
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_4bit=True,  # Enables 4-bit quantization
    device_map="auto"
).to(torch.device("cuda"))  # Load model onto GPU
```

```python
from datasets import load_dataset

dataset = load_dataset("Shekswess/llama3_medquad_instruct_dataset")

# Check column names
print(dataset["train"].column_names)

# Print a sample
print(dataset["train"][0])
```

```python
# Check if pad_token exists
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # Set pad_token as eos_token
```

```python
def preprocess_function(examples):
    # Prepare input texts (Question + Answer)
    input_texts = ["Question: " + q + " Answer: " + a for q, a in zip(examples["input"], examples["output"])]

    # Tokenize and include labels (shift input_ids to be the labels for language modeling)
    outputs = tokenizer(
        input_texts,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="pt"  # Ensure the correct tensor format
    )

    # Shift labels to the right for autoregressive models (causal LM)
    outputs['labels'] = outputs['input_ids'].clone()
    return outputs

tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
from peft import LoraConfig, get_peft_model

config = LoraConfig(
    r=8,
    lora_alpha=32,
    lora_dropout=0.1,
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, config)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="llama3_medquad_finetuned",
    per_device_train_batch_size=1,  # Prevent CUDA OOM
    gradient_accumulation_steps=8,  # Mimics batch size of 8
    num_train_epochs=3,
    fp16=True,  # Half precision for memory efficiency
    save_strategy="epoch"
)

trainer = Trainer(
    model=model,
    train_dataset=tokenized_dataset["train"],
    args=training_args,
)
trainer.train()
```
